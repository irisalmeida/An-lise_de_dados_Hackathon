```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```
#Planilha de dados qualitativos
data_intern=pd.read_csv("/content/drive/MyDrive/Provi Hack/Teste-1 - Página1.csv")
```

```
#Tirando os valores nulos da coluna
summary = data_intern.dropna(subset=['summary'], axis=0)['summary']
```

```
#Contando frequência de repetição das palavras

#pesquisa qualitativa para quantitativa
all_summary = " ".join(s for s in summary)
```

```
print("Quantidade de Palavras: {}".format(len(all_summary)))
```

```
#lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(["pouco","Júnior","requerimentos","exigidas","exigem","encontrar","início","altos","contratar", "salário","dificuldade","técnico","inscrevi","absurda","das","exigencia","pleno", "relação","mas", "ter","exigida","muita", "gupy","pra","tem", "quer", "kenoby", "estar", "chamada","só","ninguém","para", "devido", "inicial", "tempo","que", "na", "não","recrutador" ,"sobre", "bem","ser","colocar", "assunto","bootcamp", "algumas","falta", "primeiro", "preparo", "vários", "daquilo","afins", "passa","um","dificulta","realmente", "pode","data","claro", "ou","ele", "te", "uma","perda","várias","sim","sabe", "deixa", "falando","pessoais" ,"aquele", "nenhuma","muito" , "está", "não" "meu", "em", "você", "de", "ao", "os","se", "gama","meu", "nem"])
```
